```python
import requests, pandas as pd, datetime
from io import StringIO

MAP_KEY = "02e2883b9c4fe92c15cb25402e4cc584"
BBOX = "-124.48,32.53,-114.13,42.01"

start = datetime.date(2020, 8, 16)
end   = datetime.date(2020, 11, 12)
all_dfs = []

current = start
while current <= end:
    chunk_end = min(current + datetime.timedelta(days=4), end)
    days = (chunk_end - current).days + 1
    date_str = current.strftime("%Y-%m-%d")
    
    url = f"https://firms.modaps.eosdis.nasa.gov/api/area/csv/{MAP_KEY}/MODIS_SP/{BBOX}/{days}/{date_str}"
    print(f"Fetching {date_str} → {chunk_end}...", end=" ")
    
    r = requests.get(url)
    df = pd.read_csv(StringIO(r.text))
    print(f"{len(df)} detections")
    all_dfs.append(df)
    current += datetime.timedelta(days=5)

combined = pd.concat(all_dfs, ignore_index=True)
combined.to_csv("data/firms_august_complex.csv", index=False)
print(f"\nDone — {len(combined)} total fire detections")
combined.head()
```
